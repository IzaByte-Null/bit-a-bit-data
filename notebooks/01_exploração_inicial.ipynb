```python
#Célula 1: Carregamento e Estilo

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
sns.set_theme(style="dark", rc={
    "axes.facecolor": "#000000",
    "figure.facecolor": "#000000",
    "grid.color": "#1A1A1A",
    "axes.edgecolor": "#333333",
    "text.color": "#E0E0E0",
    "xtick.color": "#B0B0B0",
    "ytick.color": "#B0B0B0"
})
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
#Célula 2: Gráfico de Ranking (Top Players)

df_ranking = pd.read_csv('../data/raw/ranking_global.csv')
df_ranking['score_acumulado'] = df_ranking['score_acumulado'].fillna(0)

sns.barplot(
    data=df_ranking.head(5), 
    x='username', 
    y='score_acumulado', 
    palette='Purples_r',
    hue='username',
    legend=False
)
plt.title('🏆 Top 5 - Engajamento no Quiz Bit-a-Bit', color='#BB86FC', pad=20)
plt.show()
```

```python
#Célula 3: Observabilidade (Horários de Pico)

df_obs = pd.read_csv('../data/raw/observabilidade_bot.csv')

sns.countplot(
    data=df_obs, 
    x='hora_do_dia', 
    color='#4B0082',
    edgecolor='#BB86FC'
)
plt.title('🔃 Fluxo de Requisições por Hora (Carga do Servidor)', color='#BB86FC', pad=20)
plt.xlabel('Hora do Dia', color='#B0B0B0')
plt.ylabel('Total de Interações', color='#B0B0B0')
plt.show()
```
