```python
#Célula 1: Carregamento e Estilo

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
#Célula 2: Gráfico de Ranking (Top Players)

df_ranking = pd.read_csv('../data/raw/ranking_global.csv')


df_ranking['score_acumulado'] = df_ranking['score_acumulado'].fillna(0)

sns.barplot(data=df_ranking.head(5), x='username', y='score_acumulado', palette='viridis')
plt.title('🏆 Top 5 - Engajamento no Quiz Bit-a-Bit')
plt.show()
```

```python
#Célula 3: Observabilidade (Horários de Pico)

df_obs = pd.read_csv('../data/raw/observabilidade_bot.csv')

sns.countplot(data=df_obs, x='hora_do_dia', color='cyan')
plt.title('🔃 Fluxo de Requisições por Hora (Carga do Servidor)')
plt.xlabel('Hora do Dia')
plt.ylabel('Total de Interações')
plt.show()
```
